```python
df_items.head()
```

```python
bow = CountVectorizer(ngram_range=(1, 1), max_df=10, min_df=2)
items_bow = bow.fit_transform(df_items["item_name_translated"])
tokens = bow.get_feature_names()
df1 = pd.DataFrame(data=items_bow.toarray(), columns=tokens)
df1
```

```python
bow_2grams = CountVectorizer(ngram_range=(1, 2), max_df=10, min_df=2, dtype=np.int8)
items_bow_2grams = bow_2grams.fit_transform(df_items["item_name_translated"])
tokens_2grams = bow_2grams.get_feature_names()
df_2grams = pd.DataFrame(data=items_bow_2grams.toarray(), columns=tokens_2grams)
df_2grams
```

```python
tfidf = TfidfVectorizer(ngram_range=(1,1), max_df=10, min_df=2)
items_tfidf = tfidf.fit_transform(df_items["item_name_translated"])
tokens_tfidf = tfidf.get_feature_names()
df3 = pd.DataFrame(data=items_tfidf.toarray(), columns=tokens_tfidf)
df3
```

```python
tfidf_2 = TfidfVectorizer(ngram_range=(1,2), max_df=10, min_df=2)
items_tfidf_2 = tfidf_2.fit_transform(df_items["item_name_translated"])
tokens_tfidf_2 = tfidf_2.get_feature_names()
df4 = pd.DataFrame(data=items_tfidf_2.toarray(), columns=tokens_tfidf_2)
df4
```
