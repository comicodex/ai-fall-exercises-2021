# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
img = cv2.imread("img/blue-red-flowers.png")
```

```python
bgr_img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(bgr_img);
```

```python
hist_b = cv2.calcHist([bgr_img], [0], None, [255], [0, 255])
plt.plot(hist_b, color="b");
```

```python
hist_g = cv2.calcHist([bgr_img], [1], None, [255], [0, 255])
plt.plot(hist_g, color="g");
```

```python
hist_r = cv2.calcHist([bgr_img], [2], None, [255], [0, 255])
plt.plot(hist_r, color="r");
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img);
```

```python
hist_gray = cv2.calcHist([gray_img], [0], None, [255], [0, 255])
plt.plot(hist_gray, color="k");
```

[Hint](# "cv2.calcHist") 


```python
# Now try to plot all 3 channels at the same time, does the histogram give you an idea of what the image looks like?
for i, col in enumerate(['b', 'g', 'r']):
    hist = cv2.calcHist([bgr_img], [i], None, [255], [0, 255])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])
    
plt.show()
    
```
