# Transformations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Translations

```python
# Load an image
img = cv2.imread("img/chameleon.jpg")
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img);
```

```python
# Store the height and width of the image (you will need them later)
height, width = img.shape[:2]
height, width
```

```python
# Create the translation Matrix
translation_matrix = np.float32([[1, 0, 25], [0, 1, 50]])
```

```python
# Apply the translation to the image using cv2.warpAffine and display the new image
img_translated = cv2.warpAffine(rgb_img, translation_matrix, (width, height))
plt.imshow(img_translated);
```

### Rotation

```python
# Create a rotation matrix to rotate the image by 45 degrees (using the center as the pivot point) 
x_center = width / 2
y_center = height / 2
angle_rotation = 45
scale_factor = 1
rotation_matrix = cv2.getRotationMatrix2D((x_center, y_center), angle_rotation, scale_factor)
```

[Hint](# "cv2.getRotationMatrix2D") 


```python
# Apply the rotation matrix to the image and display the rotated image
img_rotated = cv2.warpAffine(rgb_img, rotation_matrix, (width, height))
plt.imshow(img_rotated);
```

### Scaling images

#### Interpolation methods

* INTER_NEAREST - a nearest-neighbor interpolation (Fastest)
* INTER_LINEAR - a bilinear interpolation (used by default) (good for upsampling)
* INTER_AREA - resampling using pixel area relation. It may be a preferred method for image decimation, as it gives moire-free results. But when the image is zoomed, it is similar to the INTER_NEAREST method. (Good for downsampling)
* INTER_CUBIC - a bicubic interpolation over 4x4 pixel neighborhood (Better)
* INTER_LANCZOS4 - a Lanczos interpolation over 8x8 pixel neighborhood (Best one)

#### Downsampling

```python
# Load and image and downsample it by half, try all the interpolation methods
# which one renders the best result?
new_width = width // 2
new_height = height // 2
img_resized = cv2.resize(rgb_img, (new_width, new_height), interpolation=cv2.INTER_NEAREST)
plt.imshow(img_resized);
```

```python
img_resized2 = cv2.resize(rgb_img, (new_width, new_height), interpolation=cv2.INTER_LINEAR)
plt.imshow(img_resized2);
```

```python
img_resized3 = cv2.resize(rgb_img, (new_width, new_height), interpolation=cv2.INTER_AREA)
plt.imshow(img_resized3);
```

```python
img_resized4 = cv2.resize(rgb_img, (new_width, new_height), interpolation=cv2.INTER_CUBIC)
plt.imshow(img_resized4);
```

```python
img_resized5 = cv2.resize(rgb_img, (new_width, new_height), interpolation=cv2.INTER_LANCZOS4)
plt.imshow(img_resized5);
```

#### Upsampling

```python
# Load and image and upsample it by double its size, try all the interpolation methods 
# which one renders the best result?
new_up_width = width * 2
new_up_height = height * 2
new_img = cv2.resize(rgb_img, (new_up_width, new_up_height), interpolation=cv2.INTER_NEAREST)
plt.imshow(new_img);
```

```python
new_img2 = cv2.resize(rgb_img, (new_up_width, new_up_height), interpolation=cv2.INTER_LINEAR)
plt.imshow(new_img2);
```

```python
new_img3 = cv2.resize(rgb_img, (new_up_width, new_up_height), interpolation=cv2.INTER_AREA)
plt.imshow(new_img3);
```

```python
new_img4 = cv2.resize(rgb_img, (new_up_width, new_up_height), interpolation=cv2.INTER_CUBIC)
plt.imshow(new_img4);
```

```python
new_img5 = cv2.resize(rgb_img, (new_up_width, new_up_height), interpolation=cv2.INTER_LANCZOS4)
plt.imshow(new_img5);
```
