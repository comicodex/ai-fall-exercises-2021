# Computer Vision Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#plt.rcParams['image.cmap'] = 'gray'
%matplotlib inline
```

```python
# Load an image and apply all types of thresholding that openCV offers (like thresh_binary, thresh_trunc, tozero, etc)
# Display the original image as well as the thresholded images
img = cv2.imread("img/disney.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 8))
plt.imshow(img_rgb);
```

```python
image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh1 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY) 
plt.figure(figsize=(10, 8))
plt.imshow(thresh1, cmap="gray");
```

```python
ret, thresh2 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY_INV)
plt.figure(figsize=(10, 8))
plt.imshow(thresh2, cmap="gray");
```

```python
ret, thresh3 = cv2.threshold(image, 120, 255, cv2.THRESH_TRUNC)
plt.figure(figsize=(10, 8))
plt.imshow(thresh3, cmap="gray");
```

```python
ret, thresh4 = cv2.threshold(image, 120, 255, cv2.THRESH_TOZERO)
plt.figure(figsize=(10, 8))
plt.imshow(thresh4, cmap="gray");
```

```python
ret, thresh5 = cv2.threshold(image, 120, 255, cv2.THRESH_TOZERO_INV)
plt.figure(figsize=(10, 8))
plt.imshow(thresh5, cmap="gray");
```
