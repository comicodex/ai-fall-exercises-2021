# Computer Vision Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Have a look at the 2 images below
text = cv2.imread('img/t1.png')
plt.figure(figsize = (12,10))
plt.imshow(text, cmap='gray');
```

```python
# Apply threshold to the images above to make them look better, try to find the best one (THEY ALL LOOK THE SAME!)
text_copy = text.copy()
gray_text = cv2.cvtColor(text_copy, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray_text, 127, 255, cv2.THRESH_TOZERO)
plt.figure(figsize = (12,10))
plt.imshow(gray_text, cmap='gray');
```

```python
# Try to apply adaptiveThreshold and see if that helps, use both ADAPTIVE_THRESH_MEAN_C and ADAPTIVE_THRESH_GAUSSIAN_C
thresh_mean = cv2.adaptiveThreshold(gray_text, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 12) 
plt.figure(figsize = (12,10))
plt.imshow(thresh_mean, cmap="gray");
```

```python
thresh_gaussian = cv2.adaptiveThreshold(gray_text, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 19, 7) 
plt.figure(figsize = (12,10))
plt.imshow(thresh_gaussian, cmap="gray");
```

```python
# Now apply Otsu's thresholding to the same images, always make sure to start with a copy of the original image
ret, otsu = cv2.threshold(gray_text, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU) 
plt.figure(figsize = (12,10))
plt.imshow(otsu, cmap="gray");
```

```python
# Now try to first apply Gaussian Blur to the image and then Otsu thresholding
blur = cv2.GaussianBlur(gray_text, (5, 5), 0)
ret, otsu2 = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU) 
plt.figure(figsize = (12,10))
plt.imshow(otsu2, cmap="gray");
```

```python
text_2 = cv2.imread('img/t2.png')
plt.figure(figsize = (12,10))
plt.imshow(text_2, cmap='gray')
```

```python
# Apply threshold to the images above to make them look better, try to find the best one 
text_2_copy = text_2.copy()
gray_text_2 = cv2.cvtColor(text_2_copy, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray_text_2, 127, 255, cv2.THRESH_BINARY)
plt.figure(figsize = (12,10))
plt.imshow(gray_text_2, cmap='gray');
```

```python
# Try to apply adaptiveThreshold and see if that helps, use both ADAPTIVE_THRESH_MEAN_C and ADAPTIVE_THRESH_GAUSSIAN_C
mean_thresh = cv2.adaptiveThreshold(gray_text_2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 19, 7) 
plt.figure(figsize = (12,10))
plt.imshow(mean_thresh, cmap="gray");
```

```python
gauss_thresh = cv2.adaptiveThreshold(gray_text_2, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2) 
plt.figure(figsize = (12,10))
plt.imshow(gauss_thresh, cmap="gray");
```

```python
# Now apply Otsu's thresholding to the same images, always make sure to start with a copy of the original image
ret, otsu2 = cv2.threshold(gray_text_2, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU) 
plt.figure(figsize = (12,10))
plt.imshow(otsu2, cmap="gray");
```

```python
# Now try to first apply Gaussian Blur to the image and then Otsu thresholding
gauss_blur = cv2.GaussianBlur(gray_text_2, (5, 5), 0)
ret, otsu3 = cv2.threshold(gauss_blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU) 
plt.figure(figsize = (12,10))
plt.imshow(otsu3, cmap="gray");
```
