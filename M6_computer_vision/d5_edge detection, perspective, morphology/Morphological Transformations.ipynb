# Computer Vision Morphological Transformations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Dilation

```python
# Create a simple 3 by 3 kernel of ones with numpy
kernel = np.ones((3, 3), dtype="uint8")
```

```python
img = cv2.imread('img/mare.jpg')
img_copy = img.copy()
gray_img = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10, 8)) 
plt.imshow(gray_img, cmap="gray");
```

```python
# Load an image, threshold it (binary_inverse) and apply dilation with openCV, try different number of iterations
ret, thresh = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY_INV)
dilation_img = cv2.dilate(thresh, kernel, iterations = 1)
plt.figure(figsize=(10, 8)) 
plt.imshow(dilation_img, cmap="gray");
```

```python
# Try with bigger kernels (5,5) and (7,7) or bigger and check what is the difference
kernel5 = np.ones((5, 5), dtype="uint8")
dilation_img5 = cv2.dilate(thresh, kernel5, iterations = 1)
plt.figure(figsize=(10, 8)) 
plt.imshow(dilation_img5, cmap="gray");
```

```python
kernel7 = np.ones((7, 7), dtype="uint8")
dilation_img7 = cv2.dilate(thresh, kernel7, iterations = 2)
plt.figure(figsize=(10, 8)) 
plt.imshow(dilation_img7, cmap="gray");
```

[Hint](# "cv2.dilate") 

```python
# Repeat the operation but this time create a special kernel with cv2.getStructuringElement, try different elements
# For example cv2.MORPH_RECT, cv2.MORPH_ELLIPSE and cv2.MORPH_CROSS, etc.
# Compare how the different kernels will affect the image
rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
rect_dilation = cv2.dilate(thresh, rect_kernel)
plt.figure(figsize=(10, 8)) 
plt.imshow(rect_dilation, cmap="gray");
```

```python
# cv2.MORPH_ELLIPSE 
ellipse_kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
ellipse_dilation = cv2.dilate(thresh, ellipse_kernel, iterations = 2)
plt.figure(figsize=(10, 8)) 
plt.imshow(ellipse_dilation, cmap="gray");
```

```python
# cv2.MORPH_CROSS
cross_kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (7, 7))
cross_dilation = cv2.dilate(thresh, cross_kernel, iterations = 3)
plt.figure(figsize=(10, 8)) 
plt.imshow(cross_dilation, cmap="gray");
```

## Erosion

```python
# Using the same kernels, apply Erosion to the thresholded image, make sure to start with a clean image
#rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
rect_erosion = cv2.erode(thresh, rect_kernel)
plt.figure(figsize=(10, 8)) 
plt.imshow(rect_erosion, cmap="gray");
```

```python
# cv2.MORPH_ELLIPSE 
ellipse_erosion = cv2.erode(thresh, ellipse_kernel)
plt.figure(figsize=(10, 8)) 
plt.imshow(ellipse_erosion, cmap="gray");
```

```python
# cv2.MORPH_CROSS 
cross_erosion = cv2.erode(thresh, cross_kernel)
plt.figure(figsize=(10, 8)) 
plt.imshow(cross_erosion, cmap="gray");
```

[Hint](# "cv2.erode") 

## Morphological Transformations: Opening / Closing

```python
# Apply opening to the thresholded image, experiment with different iterations 
open_img = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, rect_kernel, iterations= 2)
plt.figure(figsize=(10, 8)) 
plt.imshow(open_img, cmap="gray");
```

[Hint](# "cv2.morphologyEx with cv2.MORPH_OPEN ") 

```python
# Apply closing to the thresholded image, experiment with different iterations 
close_img = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, ellipse_kernel, iterations= 1)
plt.figure(figsize=(10, 8)) 
plt.imshow(close_img, cmap="gray");
```
