```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Corners

```python
image = cv2.imread('img/chess.jpg')
plt.imshow(image);
```

```python
# We will work with the image above
# Convert it to grayscale
image_copy = image.copy()
image_gray = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)

# Detect the corners using the cornerHarris
image_gray = np.float32(image_gray) # as cornerHarris method takes float values
corners = cv2.cornerHarris(image_gray, 5, 3, 0.01)
corners = cv2.dilate(corners, None) # to get better results
image_copy[corners > 0.01 * corners.max()] = [0, 255, 0]

# Display the image with the detected corners 
plt.figure(figsize=(12, 8))
plt.imshow(image_copy);
```

```python
img = cv2.imread('img/chess.jpg')
img_copy = img.copy()
# Convert img to grayscale
img_gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)

# Detect the corners using goodFeaturesToTrack
max_corners = 100
distance = 10
quality = 0.01
corners = cv2.goodFeaturesToTrack(img_gray, max_corners, quality, distance)

# Draw the corners on top of img and show it on screen (make sure the corners are visible!)
for corner in corners:
    x, y = corner[0] 
    x = int(x) 
    y = int(y) 
    cv2.rectangle(img_copy, (x-5, y-5), (x+5, y+5), (0, 255, 0), -1)
plt.figure(figsize=(12, 8))   
plt.imshow(img_copy);
```
