### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#%matplotlib qt
```

### To edit an image making a copy of it is always a good idea!

```python
# Load an image
# And make a copy of the original image to avoid manipulating the original image
img = cv2.imread("img/field.jpg")
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
copy_img = rgb_img.copy()
```

[Hint](# "Numpy .copy function") 

## Drawing

### Drawing lines

```python
# Draw some lines on top of the image 

cv2.line(copy_img, (20, 10), (200, 10), (0, 255, 0), 2)
cv2.line(copy_img, (50, 60), (250, 10), (0, 0, 0), 1)
plt.imshow(copy_img);
```

[Hint](# "cv2.line function") 

### Circles

```python
# Draw some circles on top of the image
cv2.circle(copy_img, (20, 120), 20, (255, 255, 0), 2)
plt.imshow(copy_img);
```

[Hint](# "cv2.circle") 

### Ellipses

```python
# Draw some elipses on top of the image
cv2.ellipse(copy_img, (150, 100), (50, 10), 0, 0, 360, (0, 0, 255), 1)
plt.imshow(copy_img);
```

[Hint](# "cv2.ellipse") 

### Poligons

```python
# Lets draw some poligons, first declare the points of the poligon
points = np.array([[180, 95], [260, 60], [260, 100], [180, 110]])
```

[Hint](# "Example: np.array([[10,5],[120,130],[170,120],[250,210]], np.int32)") 

```python
# Draw the poligon on top of the image
cv2.polylines(copy_img, [points], True, (255, 0, 0), 2)
plt.imshow(copy_img);
```

[Hint](# "cv2.polylines(img_copy,[pts],False,(255,0,0),3)")  

### Add an overlay on top of an image

```python
# Load an image and create a copy of it you can call it overlay
# Create a rectangle with a fill color (-1) on top of the overlay image
# Combine both images setting some opacity on the overlay
# Add some text on top of the overlay and show the image on screen
```

[Hint](# "cv2.addWeighted")  

```python
new_img = cv2.imread("img/buildings.jpg")
rgb_new_img = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
overlay = rgb_new_img.copy()
cv2.rectangle(overlay, (0, 0), (100, 100), (0, 0, 0), -1)

opacity = 0.5
op_img = cv2.addWeighted(overlay, opacity, rgb_new_img, 1- opacity, 0)

cv2.putText(op_img, "Good night", (2, 50), cv2.FONT_HERSHEY_PLAIN, 1, (255, 255, 255), 0, cv2.LINE_AA)
plt.figure(figsize=(10, 8))
plt.imshow(op_img);
```
