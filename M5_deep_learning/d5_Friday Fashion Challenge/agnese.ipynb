```python
from PIL import Image
import os
from matplotlib import image
from matplotlib import pyplot
```

```python
image = Image.open("/work/images/one.jpg");

print(image.format)
print(image.mode)
print(image.size)

image.show()
```

```python
print(image.size)
img_resized = image.resize((28,28))
print(img_resized.size)
```

```python
# convert the image to grayscale
gs_image = img_resized.convert(mode='L')
gs_image.save('one_grayscale.jpg')
image2 = Image.open('one_grayscale.jpg')
image2.show()
```
